Find records that has disagreement among annotators 

```python
def has_disagreement(label_list):
    label_list = eval(label_list)
    return 'YES' in label_list and 'NO' in label_list

disagreement_records = data[data['labels_task1'].apply(has_disagreement)]
disagreement_records=disagreement_records[['id','tweet','number_annotators','annotators','gender_annotators','age_annotators','labels_task1',"Total_vote","Total_age1","Total_age2","Total_age3","majority_vote"]]
disagreement_records.to_excel('EXIST_2023/Age_Complete_disagreement_records.xlsx', index=False)
```
